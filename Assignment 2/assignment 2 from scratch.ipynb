```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(5,10)
        self.fc2 = nn.Linear(10,1)

    def forward(self,x):
        x = self.fc1(x)
        x = torch.sigmoid(x)
        x = self.fc2(x)
        x = torch.sigmoid(x)
        return x

model = Net()
print(model)

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)

allloss = []

for epoch in range(100):
    outputs = model(X)
    loss = criterion(outputs,Y)
    loss.backward()
    optimizer.step()
    allloss.append(loss.item())

import matplotlib.pyplot as plt
plt.plot(allloss)
plt.show()


import sys
sys.exit(0)
```

```python
# Multilayer Perceptron from scratch

weights = [-0.1, 0.20, -0.23, -0.1, 0.20, -0.23, -0.1, 0.20, -0.23] # Weights are randomly chosen

# Using Sigmoid activation function
import math
def sigmoid(z):
    if(z<-100): 
        return 0
    if(z>100):
        return 1
    return 1.0/math.exp(-z)

# Make the first hidden layer of the multilayer perceptron
def firstLayer(row,weights):
    activation_1 = weights[0]*1 # The bias = 1
    activation_1 += weights[1]*row[0]
    activation_1 += weights[2]*row[1]

    activation_2 = weights[3]*1
    activation_2 += weights[4]*row[2]
    activation_2 += weights[5]*row[3]
    return sigmoid(activation_1),sigmoid(activation_2)

# Make the second hidden layer of the multilayer perceptron
def secondLayer(row,weights):
    activation_3 = weights[6]
    activation_3 += weights[7]*row[0]
    activation_3 += weights[8]*row[1]
    return sigmoid(activation_3)

# Then we want to do the actual prediction
def predict(row,weights): 
    input_layer = row
    first_layer = firstLayer(input_layer,weights) 
    second_layer = secondLayer(first_layer,weights) 
    return second_layer,first_layer

for d in training_dataset:
    print(predict(d,weights)[0],d[-1])
```
